```python
import numpy as np
import pandas as pd

# Load the dataset
data = pd.read_csv('hw_1_dataset_1-1.csv')
X = data.iloc[:, 0]  # Assuming the first column is the feature
y = data.iloc[:, 1]  # Assuming the second column is the target variable

# Initialize parameters
w0 = 0.0  # Intercept
w1 = 0.0  # Slope

# Learning rate
alpha = 0.01

# Number of iterations
iterations = 1000

# Number of examples
m = len(X)

# Gradient Descent
for i in range(iterations):
    # Predicted value
    y_pred = w0 + w1 * X
    # Compute gradients
    dw0 = (1/m) * np.sum(y_pred - y)
    dw1 = (1/m) * np.sum((y_pred - y) * X)
    # Update parameters
    w0 = w0 - alpha * dw0
    w1 = w1 - alpha * dw1

print(f'w0: {w0}')
print(f'w1: {w1}')
```

```python
import matplotlib.pyplot as plt

# Plotting the original data points
plt.scatter(X, y, color='blue', label='Data points')

# Plotting the prediction line
x_values = np.linspace(min(X), max(X), 100)  # Generate 100 points for a smooth line
y_values = w0 + w1 * x_values
plt.plot(x_values, y_values, color='red', label='Prediction line')

# Labeling the plot
plt.xlabel('X')
plt.ylabel('y')
plt.title('Data and Prediction Line')
plt.legend()

# Display the plot
plt.show()
```
