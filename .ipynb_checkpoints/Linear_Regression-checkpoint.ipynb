**Linear Regresion (non-matrix format)**

```python
import numpy as np
from matplotlib import pyplot as plt

w0_ = 35
w1_ = 7

x = np.arange(0, 0.5, 0.01)
x = np.expand_dims(x, axis = 1)
y = w0_ + w1_*x

fig, ax = plt.subplots(1, 1, figsize = (5,5))
ax.scatter(x,y, color='r')
```

### Add Noise to data

```python
noise = np.random.normal(loc = 0, scale=3, size=x.shape)
```

```python
y = y + noise
print(x.shape)
print(y.shape)
print(noise.shape)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (5,5))
ax.scatter(x,y, color='r')
```

# Uni-variate linear regression

### Hyperparameters

```python
epochs = 10001
alpha = 0.001 #Learning rate
epsilon_w0 = 0.001 #Tolerance for breaking the loop
epsilon_w1 = 0.001
M = x.shape[0]
```

### Initialize parameters

```python
w0 = abs(np.random.normal())
w1 = abs(np.random.normal())

print(w0, w1)
```

### Gradient descent

#### Hypothesis using an anonymous function/lambda function 

```python
hw = lambda w0, w1, x: (w0 + (w1*x)) #hypothesis
```

```python
for n in range(0, epochs):
    # Cost function
    J = (1/2*M) * (sum(hw(w0, w1, x) - y)**2)
    
    # Calculate the partial derivatives of the cost function
    dfJ_w0 = (1/M) * sum(hw(w0, w1, x) - y)
    dfJ_w1 = (1/M) * sum((hw(w0, w1, x) - y) * x) 
    
    # Update parameters using gradients
    w0_old = w0
    w1_old = w1
    w0 = w0 - (alpha * dfJ_w0)
    w1 = w1 - (alpha * dfJ_w1)
    
    # Check for convergence
    if((abs(w0_old - w0) < epsilon_w0) & (abs(w1_old - w1) < epsilon_w1)):
       print(n)
       break
```

```python
print(w1, w0)
```

```python
y_pred = w0 + w1*x
```

```python
fig, ax = plt.subplots(1,1, figsize=(5,5))
ax.scatter(x,y, color='b')
ax.plot(x,y_pred,color='r')
```
