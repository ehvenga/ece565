# Uni-variate Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('hw_1_dataset_1-1.csv')
x = data['x'].values # Assuming the first column is the feature
y = data['y'].values  # Assuming the second column is the target variable
```

### Initialize Parameters

```python
w0 = 0
w1 = 0
epsilon_w0 = 0.001 #Tolerance for breaking the loop
epsilon_w1 = 0.001
alpha = 0.01 # learning rate
epochs = 1000 # Iterations
M = x.shape[0]
```

```python
hw = lambda w0, w1, x: (w0 + (w1*x)) #hypothesis
```

```python
for n in range(0, epochs):
    # Cost function
    J = (1/2*M) * (sum((hw(w0, w1, x) - y)**2))
    
    # Calculate the partial derivatives of the cost function
    dfJ_w0 = (1/M) * sum(hw(w0, w1, x) - y)
    dfJ_w1 = (1/M) * sum((hw(w0, w1, x) - y) * x) 
    
    # Update parameters using gradients
    w0_old = w0
    w1_old = w1
    w0 = w0 - (alpha * dfJ_w0)
    w1 = w1 - (alpha * dfJ_w1)
    
    # Check for convergence
    if((abs(w0_old - w0) < epsilon_w0) & (abs(w1_old - w1) < epsilon_w1)):
       print(n)
       break
```

```python
print('w0:',w0,'\nw1:',w1)
```

```python
plt.scatter(x, y, color='b', label='Data points')
x_values = x
y_values = w0 + w1 * x_values
plt.plot(x_values, y_values, color='r', label='Prediction line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data and Prediction Line')
plt.legend()
plt.show()
```
